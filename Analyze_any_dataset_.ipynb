```
import pandas as pd
from pandas_profiling import ProfileReport
df = pd.read_csv("housing.csv")
print(df)
```

```

profile = ProfileReport(df)
# this below lines show the upto the point data
#profile = ProfileReport(df,minimal=true)
profile.to_file(output_file="housing.html")
```

```

profile = ProfileReport(df)
# this below lines show the upto the point data, basically for large data sets
profile = ProfileReport(df,minimal=True)
profile.to_file(output_file="minimal.html")
```

```

profile = ProfileReport(df)
# this below lines show the report with the title 
profile = ProfileReport(df, title="Pandas Profiling Report")
profile.to_file(output_file="with_title.html")
```

```

profile = ProfileReport(df)
# this below lines show the report with exploring into the deep os the data 
profile = ProfileReport(df, title='Pandas Profiling Report', explorative=True)
profile.to_file(output_file="explore_deep.html")
```

```
#we can obtain the data in json
# As a string
json_data = profile.to_json(df)

# As a file
profile.to_file("your_report.json")
```
